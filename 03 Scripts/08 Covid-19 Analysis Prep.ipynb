## 2. Data Prep

```python
#Making pre and post pandemic variables for norm_sale_amount, norm_assessed_value, and norm_sales_ratio

#Pre-pandemic values
df.loc[df['in_pandemic'] == 0, 'pre_sale_amount'] = df['norm_sale_amount']
df.loc[df['in_pandemic'] == 0, 'pre_assessed_value'] = df['norm_assessed_value']
df.loc[df['in_pandemic'] == 0, 'pre_sales_ratio'] = df['norm_sales_ratio']

#Pre-pandemic values
df.loc[df['in_pandemic'] == 1, 'post_sale_amount'] = df['norm_sale_amount']
df.loc[df['in_pandemic'] == 1, 'post_assessed_value'] = df['norm_assessed_value']
df.loc[df['in_pandemic'] == 1, 'post_sales_ratio'] = df['norm_sales_ratio']
```

```python
#Checking new variable

df.head(20)
```

```python
#Checking new variable

df.tail(20)
```

```python
#Creating data fram of pre and post covid mean statistics per town for sale_amount, assessed_value, and sales_ratio

df_covid = pd.DataFrame()
df_covid['pre_sale_amount'] = df.groupby('town')['pre_sale_amount'].mean()
df_covid['pre_assessed_value'] = df.groupby('town')['pre_assessed_value'].mean()
df_covid['pre_sales_ratio'] = df.groupby('town')['pre_sales_ratio'].mean()
df_covid['post_sale_amount'] = df.groupby('town')['post_sale_amount'].mean()
df_covid['post_assessed_value'] = df.groupby('town')['post_assessed_value'].mean()
df_covid['post_sales_ratio'] = df.groupby('town')['post_sales_ratio'].mean()

df_covid = df_covid.reset_index()
```

```python
#pre/post covid change variables

df_covid['changed_sale_amount'] = df_covid['post_sale_amount'] - df_covid['pre_sale_amount']
df_covid['changed_assessed_value'] = df_covid['post_assessed_value'] - df_covid['pre_assessed_value']
df_covid['changed_sales_ratio'] = df_covid['post_sales_ratio'] - df_covid['pre_sales_ratio']
```
